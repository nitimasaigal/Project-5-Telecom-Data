Task 3.4 - Using the experience metrics above, perform a k-means clustering (where k = 3) to segment users into groups of experiences and provide a brief description of each cluster. (The description must define each group based on your understanding of the data)

```python
experience_metrics = df[['Aggregate_TCP', 'Aggregate_RTT', 'Aggregate_TP']]
```

Here we have taken experience metrics as features

```python
plt.scatter(df['Aggregate_TCP'], df['Aggregate_RTT'], df['Aggregate_TP'])
plt.title('ScatterPlot for Experience Metrics')
```

Scaling

```python
from sklearn.preprocessing import MinMaxScaler
experience_scaler = MinMaxScaler()
experience_scaled = experience_scaler.fit_transform(experience_metrics)
experience_scaled
```

KMeans Clustering

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters= 3)
df['experience_cluster'] = km.fit_predict(experience_metrics)
```

```python
df['experience_cluster']
```

```python
df['experience_cluster'].unique()
```

```python
df0 = df[df.experience_cluster==0]
df1 = df[df.experience_cluster==1]
df2 = df[df.experience_cluster==2]
plt.scatter(df0['Aggregate_TCP'],df0['Aggregate_RTT'],df0['Aggregate_TP'],color = 'green')
plt.scatter(df1['Aggregate_TCP'],df1['Aggregate_RTT'],df1['Aggregate_TP'],color = 'red')
plt.scatter(df2['Aggregate_TCP'],df2['Aggregate_RTT'],df2['Aggregate_TP'],color = 'pink')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color = 'black',marker='x',label = 'centroid')
plt.xlabel('Clusters')
plt.ylabel('Experience Metrics')
plt.title('Scatterplot for Centroid of clusters')
plt.legend()
```

Calculating the value of k

```python
ssd_values = []
for i in range(1,10):
    km = KMeans(n_clusters=i)
    km.fit(experience_scaled)
    ssd_values.append(km.inertia_)
```

```python
plt.plot(range(1,10),ssd_values)
plt.title('Elbow Method for k')
```

Here value of k should be 2

Calculating Experience Labels

```python
labels = km.labels_
labels = pd.DataFrame(labels)

km_df = pd.concat([experience_metrics, labels], axis= 1)
km_df = km_df.rename(columns={0:'experience_labels'})
km_df
```

```python
experience_metrics['k = 3 experience_labels'] = km.labels_
experience_metrics.corr()['k = 3 experience_labels'].sort_values().plot(kind= 'bar')
plt.title('Correlation between Experience metrics and Experience labels')
```

```python
plt.figure(figsize = (12,5), dpi = 200)
sns.scatterplot(data= km_df, cmap = 'magma')
plt.title('ScatterPlot for Experience Clusters')
plt.xlabel('Experience Metrics')
plt.ylabel('Labels')
plt.show()
ImportWarning
```
